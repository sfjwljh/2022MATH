```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("附件.xlsx",sheet_name="表单1")
data
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
lst = []
lst_key = []
for itm in ["A","B","C"]:
    lst_key.append(itm)
    labels = data.loc[data['纹饰']==itm,['表面风化']]
    count_1 = 0
    count_2 = 0
    for label in labels.values:
        if label[0] =='风化':
            count_1=count_1+1
        else:
            count_2=count_2+1
    value = count_1/(count_1+count_2)
    lst.append(value)
lst = np.array(lst)
lst_key=np.array(lst_key)
lst_key = lst_key[np.argsort(lst)]
lst.sort()
x_pos = range(len(lst_key)) 
plt.bar(x_pos, lst, align='center', alpha=0.5)
plt.xticks(x_pos, lst_key)
plt.ylabel('风化占比')
plt.title(r'各特征值风化占比图')
plt.show()
```
